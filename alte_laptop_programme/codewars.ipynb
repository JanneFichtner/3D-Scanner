```python
import numpy as np
import pandas as pd

# Tabelle aus dem Bild als strukturierte Daten
werkstoffe = [
    "Stahl", "Al-Legierung", "Mg-Legierung", "Ti-Legierung",
    "PA6 (trocken)", "GFK-UD (60%)", "CFK-UD (60%)", "AFK-UD (60%)"
]

rho = [7.85, 2.7, 1.74, 4.5, 1.15, 1.95, 1.40, 1.35]               # [kg/dm³]
E = [210000, 73000, 45000, 108000, 2500, 40000, 270000, 70000]     # [N/mm²]
sigma_b = [500, 350, 330, 900, 80, 800, 1000, 1500]                # [N/mm²]
k = [0.1274, 0.37, 0.575, 0.222, 0.869, 0.526, 0.645, 0.752]       # 1/(rho*1000)

# DataFrame erstellen
df = pd.DataFrame({
    "Werkstoff": werkstoffe,
    "ρ [kg/dm³]": rho,
    "E [N/mm²]": E,
    "σ_B [N/mm²]": sigma_b,
    "1/ρ [dm³/kg]": k
})
#Konstanten:
F = 300
L = 500
t = 2
# Berechnungen hinzufügen
df["Knickgütezahl"] = [e**(1/3) * ki for e, ki in zip(df["E [N/mm²]"], k)]

# Funktion für dm
def berechne_dm(E):
    return ((8 * 300 * (500 ** 3)) / (3 * np.pi * E * 2)) ** (1/3)

def berechne_dmb(sigma_b):
    return ((4 * F * L) /(np.pi * sigma_b * t)) ** (1/2)

# Funktion für Masse
def berechne_m(dm, rho):
    return rho * np.pi * dm * 2 * 500 / 1000

# dm und m berechnen
df["d_m [mm]"] = df["E [N/mm²]"].apply(berechne_dm)
df["m [g]"] = [berechne_m(dm, r) for dm, r in zip(df["d_m [mm]"], rho)]
df["d_m b) [mm]"] = df["σ_B [N/mm²]"].apply(berechne_dmb)
df["Dichte Aufgabe b)"]  = [berechne_m(dm, r) for dm, r in zip(df["d_m b) [mm]"], rho)]

df
```
