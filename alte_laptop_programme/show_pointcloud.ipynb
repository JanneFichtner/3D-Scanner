```python
import numpy as np
import open3d as o3d
from pathlib import Path

PATH = r"G:\Meine Ablage\Studium\Master\3. Semester\Masterprojekt\Punktwolken_Scans_Prototyp\Maus_23102025.npz"

def load_points(npz_path: str) -> np.ndarray:
    data = np.load(npz_path)
    pts = data["P"] if "P" in data else data[list(data.files)[0]]
    return np.asarray(pts, dtype=float).reshape(-1, 3)

# 1) Punktwolke erstellen + Normals schätzen (für Poisson hilfreich)
points = load_points(PATH)
pcd = o3d.geometry.PointCloud(o3d.utility.Vector3dVector(points))
pcd.estimate_normals(
    search_param=o3d.geometry.KDTreeSearchParamHybrid(radius=5.0, max_nn=50)
)
# Optional, verbessert die Orientierung in einem konsistenten Nachbarschaftsbereich:
pcd.orient_normals_consistent_tangent_plane(30)

# 2) Poisson-Rekonstruktion (depth=8..10 ist üblicher; 2 ist sehr grob)
print("Poisson-Rekonstruktion …")
mesh, densities = o3d.geometry.TriangleMesh.create_from_point_cloud_poisson(
    pcd, depth=9
)

# 3) Dünne Randregionen per Dichte-Maske wegschneiden (sonst “Schale” um alles)
densities = np.asarray(densities)
keep = densities > np.quantile(densities, 0.02)  # untere 2% entfernen
mesh = mesh.select_by_index(np.where(keep)[0])

# 4) Mesh aufräumen + Vertex-Normalen für Shading berechnen
mesh.remove_duplicated_vertices()
mesh.remove_degenerate_triangles()
mesh.remove_non_manifold_edges()
mesh.compute_vertex_normals()      # <- entscheidend für Beleuchtung
# (Optional für Flat-Shading:)
# mesh.compute_triangle_normals()

# 5) Farbe setzen (sonst manchmal “dunkelgrau”)
mesh.paint_uniform_color([0.7, 0.75, 0.85])

# 6) Anzeigen – Backface-Culling optional ausschalten
o3d.visualization.draw_geometries(
    [mesh],
    mesh_show_back_face=True,  # sieh auch Rückseiten; hilft bei falsch orientierten Normals
    zoom=0.7,
)
```
