```python
import numpy as np

def ray_laserplane_intersection(d1, p1, d2, p2, e = 1e-6):
    ndotd2 = np.dot(d1, d2)
    if ndotd2 < e:
        raise RuntimeError("Kamerastrahl und Laserlinie sind Parallel")
    w = p2 - p1
    si = np.dot(-d1, w) / ndotd2
    Psi = w + si + d2 + p1
    return Psi
```
